1. Role of Weights in a Neuron

In an artificial neuron, weights are learnable parameters that determine the importance of each input feature.

Mathematically:

z = w1x1 + w2x2 + ... + b

Where:
w = weights
x = input features
b = bias

Weights control how strongly each input affects the output. During training, weights are updated using gradient descent to minimize the loss function.

2. What is an Activation Function?

An activation function introduces non-linearity into the model.

It is applied after computing:

z = wx + b

Then:

a = f(z)

Common activation functions:

Sigmoid (binary classification)

ReLU (deep networks)

Softmax (multi-class classification)

For logistic regression, we use Sigmoid:

σ(z) = 1/ 1+e^-z

3. Probability Distribution in Machine Learning

A probability distribution describes how probabilities are assigned to different outcomes.

In machine learning:

-Logistic Regression assumes Bernoulli distribution

-Linear Regression assumes Gaussian distribution

Probability distributions are important in:

-Maximum Likelihood Estimation

-Loss function formulation

-Bayesian learning

4. What is Gradient in Optimization?

Gradient is the derivative of the loss function with respect to model parameters.

∂L/∂w

It tells:

-Direction to update weights

-Magnitude of update

Used in:

-Gradient Descent

-Backpropagation

The goal is to move in the negative gradient direction to minimize loss.

```python
# Step 1: Import Libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Create Dataset

# Study hours
X = np.array([[1], [2], [3], [4], [5], [6]], dtype=float)

# Pass (1) / Fail (0)
y = np.array([[0], [0], [0], [1], [1], [1]], dtype=float)

print("Input Shape:", X.shape)
print("Output Shape:", y.shape)
```

```python
# Step 3: Build Logistic Regression Model

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# Step 4: Train Model

history = model.fit(X, y, epochs=200, verbose=0)

print("Training Completed")
```

```python
# Step 5: Make Predictions

predictions = model.predict(X)

print("Predictions:")
print(predictions)
```

```python
# Step 6: Visualize Results

plt.scatter(X, y, color='red')
plt.plot(X, predictions, color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Probability of Passing")
plt.title("Logistic Regression using TensorFlow")
plt.show()
```
